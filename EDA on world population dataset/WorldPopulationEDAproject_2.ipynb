```python
data[data["Continent"]=="Europe"][["Country", "2022 Population"]].sort_values(by="2022 Population", ascending=False, ignore_index=True)[:5]
```

```python
europe = data[data["Continent"]=="Europe"][["Country", "2022 Population"]].sort_values(by="2022 Population", ascending=False, ignore_index=True)[:5]
```

```python
sns.barplot(data=europe, x="Country", y="2022 Population")
```

```python
for x in data["Continent"].unique():
  temp = data[data["Continent"]==x][["Country", "2022 Population"]].sort_values(by="2022 Population", ascending=False, ignore_index=True)[:5]
  plt.figure(figsize=(12,4))
  plt.title(f"Top 5 countries based on 2022 Popluation in {x}")
  sns.barplot(data=temp, x="Country", y="2022 Population", palette="Set2")
  plt.show()
```

```python
diff = data.groupby("Continent")[["1970 Population", "2022 Population"]].sum().sort_values(by="Continent", ascending=True).reset_index()
```

```python
diff
```

```python
plt.figure(figsize=(16,8))
plt.plot(diff["Continent"], diff["1970 Population"], label="1970", color="red", marker="o")
plt.plot(diff["Continent"], diff["2022 Population"], label="2022", color="blue", marker="o")
plt.xlabel("Continent")
plt.ylabel("Population")
plt.grid()
plt.legend()
plt.show()
```
