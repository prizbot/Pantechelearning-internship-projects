```python
dataset["Month"]=dataset["Order Date"].str[:2]
```

```python
dataset.head()
```

```python
dataset.groupby("Month")["Sales"].sum().sort_values(ascending=False)
```

## **7. Probability of Next Order Being USB-C Charging Cable**
To calculate the probability of the next order being a USB-C Charging Cable, we use the frequency of USB-C Charging Cable orders divided by the total number of orders.

```python
def probability(frequency, total_frequency):
  return frequency/total_frequency
```

```python
total = dataset["Product"].value_counts().sum()
```

```python
usb_c = dataset[dataset["Product"] == "USB-C Charging Cable"]["Quantity Ordered"].value_counts().sum()
```

```python
probability(usb_c, total)
```

This shows the probability for next people will order USB-C
Charging Cable

---



## **8. Probability of Next Order Being Other Products**
To calculate the probability of the next order being any other product, we use the frequency of each product's orders divided by the total number of orders.

```python
for x in dataset["Product"].unique():
  usb = dataset[dataset["Product"] == x]["Quantity Ordered"].value_counts().sum()
  print(x, probability(usb, total))
```

The probabilities for each product will show the likelihood of the next order being that specific product.
